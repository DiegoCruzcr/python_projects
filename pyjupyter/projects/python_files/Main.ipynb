```python
import pandas as pd
```

```python
file_read = pd.read_csv('C:\python_myprojects\pyjupyter\projects\data\wine_dataset.csv') 
```

```python
file_read.head()
```

```python
file_read['style'] = file_read['style'].replace('red', 0)
file_read['style'] = file_read['style'].replace('white', 1)
```

```python
file_read.info()
```

```python
file_read.head()
```

```python
#separando variav√©l alvo 
variable_target = file_read['style'] 
new_dataset = file_read.drop('style', axis=1)
```

```python
new_dataset.head()
```

```python
variable_target
```

```python
from sklearn.model_selection import train_test_split
#separando dados de treino e teste

variable_target_train, variable_target_test, new_dataset_train, new_dataset_test = train_test_split(variable_target, new_dataset, test_size=0.3)
```

```python
from sklearn.ensemble import ExtraTreesClassifier
#criando modelo
classifier_model = ExtraTreesClassifier()
classifier_model.fit(variable_target_train, new_dataset_train)

#imprimindo resultados:
result = classifier_model.score(variable_target_test, new_dataset_test)
result = (result * 100)
print(f'Accuracy: {result:.2f}%')
```
